```python
q1 = data["age"].quantile(.25)
q3 = data["age"].quantile(.75)
iqr = q3-q1
iqr
```

```python
df = data[(data["age"]> (q1-1.5*iqr)) & (data["age"]< (q3 +1.5*iqr))]
df
```

```python
names = df.columns[4:40]
names
```

```python
scaled_feature = df.copy()
```

```python
features = scaled_feature[names]
features
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(features.values)
```

```python
print(scaler.mean_)
```

```python
features = scaler.transform(features.values)
```

```python
scaled_feature[names] = features
```

```python
scaled_feature.head()
```

```python
scaled_feature.describe()
```

```python
data.head()
```

```python
data.describe()
```

```python
def gender_to_numeric(x):
    if x == "M":
        return 1
    if x == "F":
        return 2
    if x == "not disclosed":
        return 3
```

```python
scaled_feature["gender"]= scaled_feature["gender"].apply(gender_to_numeric)
```

```python
scaled_feature["gender"].head()
```

```python
# Building k means model
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 5, random_state = 0)
```

```python
model = kmeans.fit(scaled_feature)
```

```python
# Elbow
wcss = []
for i in range(1,20):
    kmeans = KMeans(n_clusters = i,random_state = 0)
    kmeans.fit(scaled_feature)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,20), wcss)
plt.title ("Elbow curve")
plt.xlabel("number of cluster")
plt.ylabel("wcss")
plt.show()
```

```python
print(wcss)
```

```python
# 5 cluster solution is good for us 
kmeans = KMeans(n_clusters = 5, random_state = 0)
kmeans.fit(scaled_feature)
```

```python
df["cluster"] = kmeans.labels_
```

```python
df
```

```python
df['cluster'].value_counts()
```

```python
# cluster size 
plt.figure(figsize= (12,6))
axis = sns.barplot(x = np.arange(0,5,1), y = df.groupby(["cluster"]).count()["age"].values)
x = axis.set_xlabel("number of cluster")
y = axis.set_ylabel("Number of students ")
```

```python
list(df.groupby(["cluster"]).mean()["age"].values)
```

```python
kmeans = KMeans(n_clusters = 5, random_state = 0)
kmeans.fit(scaled_feature)
pred = kmeans.fit_predict(scaled_feature)
```

```python
from sklearn.metrics import silhouette_score
score2 = silhouette_score(scaled_feature,pred)
score2
```
